```python
from estrutarq.utilitarios.informacoes import *
from random import sample

quantidade = 11
codigos = sample([cod for cod in range(10000, 90000000)], quantidade)
nomes = lista_nomes(quantidade)
telefones= lista_telefones(quantidade)

print(codigos)
print(nomes)
print(telefones)
```

```python
# Arquivo com registros de comprimento variável
from estrutarq.registro import *
from estrutarq.campo import *
from random import sample, randint

registro = RegistroBruto(
    ("codigo", CampoIntBinario()),
    ("nome", CampoCadeiaTerminador()),
    ("telefone", CampoCadeiaFixo(15)),
    ("salario", CampoRealBinario())
)

arquivo = open("dados", "wb")

for i in range(quantidade):
  registro.codigo.valor = codigos[i]
  registro.nome.valor = nomes[i]
  registro.telefone.valor = telefones[i]
  registro.salario.valor = randint(300000, 1500000)/100

  registro.escreva(arquivo)

arquivo.close()

!hd dados
```
